#### This is the complete Inventory Management Setup. Here first of all it takes number of total items the person wants to buy, accordingly the loop will run that much time.We get id numbers of every products from the barcode scanner. Then the user put how many items of that product he/she wants to buy. If the requirement is greater than the amount of the current product amount, then it will return a message"out of stock", and assume that value as '0'. At the end it gives the total amount of the bill. with the current stock of goods available at the shop 


```python
shop = []
value = []
quantity = []
total = []
remaining = []
summation = 0
print("*****************WELCOME TO OUR SHOP*****************")
print("\n")
 
n = int(input("enter how many products you want = "))
for i in range(0,n):
    
    items = int(input("Enter product ids = "))
    shop.append(items)

for i in range(0,n):
    value.append(records[shop[i]]['price'])

for i in range(0,n):
    amm = int(input("Enter amount of products = "))
    if(amm<=int(records[shop[i]]['amount'])):
        quantity.append(amm)
    else:
        print("out of stock")
        quantity.append(0)

#print(shop) 
print("\n")
print("The per piece value of your products = ",value)
print("The amount you are buying every product = ",quantity)
print("\n")
for i in range(0,len(quantity)):
    total.append(int(value[i]) * int(quantity[i]))
    
#print(total)

for ele in range(0, len(total)):
    summation = summation + total[ele]
 
 
print("GRAND TOTAL BILL : ", summation)
print("\n")
print("SHOP WITH US AGAIN")



print("remaining amount of goods after shooping")
print("\n")
for i in range(0,n):
    remaining.append(int(records[shop[i]]['amount'])-quantity[i])
print("The remaining amount of sold products = ",remaining)
print('\n')
print("The overall current stocks = ")
print('\n')
for i in range(0,n):
    records[shop[i]]['amount'] = remaining[i]
records
```
